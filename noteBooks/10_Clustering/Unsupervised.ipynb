In general, there is two fundemental clustering methodlogies : `Soft` & `Hard`.
The key difference between the two lies behind the idea of by what degree
does each member relates to its cluster. Is it binary ( 0 or 1 ), i.e it either
belongs to that cluster or NOT. Or is it probablistic, i.e it belongs to this
cluster by X%, which means it belongs to the others as well.

## Generating Blobs

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X,y = make_blobs( n_features=2, centers=3, n_samples=150, random_state=0, cluster_std=0.4)
```

```python
plt.scatter( X[:,0], X[:,1], c="white", edgecolors="black")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.grid()
plt.tight_layout()
plt.show()
```

## KNN

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans( n_clusters=3, n_init=10, max_iter=300, tol=1e-3 )
```

```python
y_km = km.fit_predict(X)
```

```python
plt.scatter(X[y_km == 0, 0], X[y_km == 0, 1], color = "crimson", edgecolors="black", alpha=0.85)
plt.scatter(X[y_km == 1, 0], X[y_km == 1, 1], color = "yellow", edgecolors="black", alpha=0.85)
plt.scatter(X[y_km == 2, 0], X[y_km == 2, 1], color = "green", edgecolors="black", alpha=0.85)
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],s=350, color="white" ,edgecolors="black", marker="*")
```

```python
'''
    To make sure we have chosen the best K, we can make the `elbow test`.
'''
```

```python
distortion = []
for i in range(1,10):
    km = KMeans( n_clusters=i, n_init=10 )
    km.fit_predict(X)
    distortion.append(km.inertia_)
```

```python
plt.plot( range(1,10), distortion )
```

```python
'''
    Other way to evaluate the quality of clustering is to make a silhoutte plot.
    This plot is based on the `Cohesion` and `Seperation` of clusters.

    * Cohesion   : The variance inside the cluster itself. How much compact it is?
    * Seperation : The variance between all clusters. How much seperated they are?

    While the former should be a small, the latter must be a big value.
'''
```

```python
k_means = KMeans( n_clusters=3, n_init=10, random_state=0 )
y_km = k_means.fit_predict(X)
```

```python
import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples
```

```python
labels = np.unique( y_km )
n_labels = labels.shape[0]
```

```python
silhoutte_vals = silhouette_samples( X=X, labels=y_km, metric="euclidean" )
```

```python
for i,c in enumerate(labels):
    c_silhoutte_vals = silhoutte_vals[y_km == c] 
    c_silhoutte_vals.sort()
    y_ax_upper += len(c_silhoutte_vals)
    color = cm.jet( float(i) / n_labels )
    plt.barh( range(y_ax_lower, y_ax_upper), c_silhoutte_vals) 
    y_ax_lower += len(c_silhouette_vals)
```

## Clusters As Hierarchial Tree

```python
'''
    * The best advantage of these is we dont have to specify the number of K
    previously.
    
    * The main idea is starting either from n clusters, n = len(sample), until 
    one big cluster has remained, or by starting from one cluster that 
    encompasses everything to n clusters.
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
columns = ['A', 'B', 'C', 'D','E']
vals = ['ID_0', 'ID_1','ID_2', 'ID_3', 'ID_4']

X = np.random.random_sample( [len(vals),len(columns)] )*10

df = pd.DataFrame( index = vals, columns=columns, data=X )
df
```

```python
from scipy.spatial.distance  import pdist
from scipy.cluster.hierarchy import linkage
link = linkage( df.values, method="complete", metric="euclidean" )
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
row_dendr = dendrogram(link, labels=vals)
```

```python
row_df = df.iloc[row_dendr['leaves'][::-1]]
```

```python
fig = plt.figure( figsize=(7,7), facecolor="white" )
axd = fig.add_axes( [.1, .1, .25, 0.7] )
row_dendr = dendrogram(link, orientation="left")
row_df = df.iloc[row_dendr['leaves'][::-1]]
axm = fig.add_axes( [.4, .15, .6, .6] )
axd.set_xticks([])
axd.set_yticks([])
for i in axd.spines.values():
    i.set_visible(False)

cax = axm.matshow(row_df, interpolation = "nearest")
fig.colorbar(cax)
```
