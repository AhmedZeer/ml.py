```python
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, [2,3]]
y = iris.target
```

```python
# Class Labels.
np.unique(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( X, y, test_size = 0.3, random_state = 1, stratify = y )
```

```python
# The np.bincount() method is used to count the occurence of "i" index in an array.
# arr = [ 0, 1, 1, 2, 4 ]
# output => [ 1, 2, 1, 0, 1] Which means there is `1` zeros, `2` ones, `1` twos etc.
print(np.bincount(y_test))
print(np.bincount(y_train))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit( x_train )
X_train_std = sc.transform( x_train )
X_test_std = sc.transform( x_test )
```

```python
from sklearn.linear_model import Perceptron
```

```python
ppn = Perceptron( eta0 = 0.1, random_state = 1 )
```

```python
ppn.fit( X_train_std, y_train )
```

```python
# Inference
y_pred = ppn.predict( X_test_std )
```

```python
# How many classes have been missclassified.
( y_pred != y_test ).sum()
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score( y_test, y_pred )
print(score)
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
```

```python
def plot_decision_regions( X, y, classifier, test_idx = None, res = 0.02 ):

    colors = ("red", "green", "blue", "yellow", "orange", "pink")
    markers = ( "o", "v", "^", "s", "<" )
    cmap = ListedColormap( colors[: len( np.unique(y))] )

    x1_min, x1_max =  X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max =  X[:, 1].min() - 1, X[:, 1].max() + 1

    xx1, xx2 = np.meshgrid( np.arange( x1_min, x1_max, res ),
                            np.arange( x2_min, x2_max, res ))

    prdct = np.array( [xx1.ravel(), xx2.ravel()] ).T
    lab = classifier.predict( prdct )
    lab = lab.reshape( xx1.shape )

    plt.contour( xx1, xx2, lab, alpha = 0.3, cmap = cmap )
    plt.xlim( xx1.min(), xx1.max() )
    plt.ylim( xx2.min(), xx2.max() )

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter( x = X[ y == cl, 0],
                     y = X[ y == cl, 1],
                     alpha = 0.8,
                     c = colors[idx],
                     marker = markers[idx],
                     label = f"Class : {cl}",
                     edgecolor = "black")
        
    if test_idx:
        X_test, Y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0], X_test[:, 1],
        c='none', edgecolor='black', alpha=1.0,
        linewidth=1, marker='o',
        s=100, label='Test set')
```
