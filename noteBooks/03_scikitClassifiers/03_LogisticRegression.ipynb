```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z):
    return 1.0 / ( np.exp(-z) + 1.0 )
```

```python
z = np.arange( -7, 7, 0.1)
```

```python
sigma_z = sigmoid(z)

plt.plot( z, sigma_z )
plt.yticks( [0.0, 0.5, 1.0 ])
plt.xticks( [-7, 0, 7 ])

plt.ylim( -0.1, 1.1)
plt.axvline(0, color="k")

ax = plt.gca()
ax.yaxis.grid(True)
ax.xaxis.grid(True)

plt.ylabel("$\sigma(z)$")
plt.tight_layout()
plt.show()
```

```python
def loss_1(z):
    return ( -np.log( sigmoid(z) ) )

def loss_0(z):
    return ( -np.log( 1 - sigmoid(z) ) )
```

```python
z = np.arange( -10, 10, 0.1 )
sigma_z = sigmoid(z)
```

```python
c0 = [ loss_0(i) for i in sigma_z ]
c1 = [ loss_1(i) for i in sigma_z ]
```

```python
plt.plot( X = sigma_z, Y = c0, c='r' )
plt.plot( X = sigma_z, Y = c1, c='b' )

plt.show()
```
