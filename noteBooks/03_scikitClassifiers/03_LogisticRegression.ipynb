```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(z):
    return 1.0 / ( np.exp(-z) + 1.0 )
```

```python
z = np.arange( -7, 7, 0.1)
```

```python
sigma_z = sigmoid(z)

plt.plot( z, sigma_z )
plt.yticks( [0.0, 0.5, 1.0 ])
plt.xticks( [-7, 0, 7 ])

plt.ylim( -0.1, 1.1)
plt.axvline(0, color="k")

ax = plt.gca()
ax.yaxis.grid(True)
ax.xaxis.grid(True)

plt.ylabel("$\sigma(z)$")
plt.tight_layout()
plt.show()
```

```python
def loss_1(z):
    return ( -np.log( sigmoid(z) ) )

def loss_0(z):
    return ( -np.log( 1 - sigmoid(z) ) )
```

```python
z = np.arange( -10, 10, 0.1 )
sigma_z = sigmoid(z)
```

```python
c0 = [ loss_0(i) for i in sigma_z ]
plt.plot( sigma_z, c0, label="y = 1" )

c1 = [ loss_1(i) for i in sigma_z ]
plt.plot( sigma_z, c1, label="y = 0" )

plt.ylim(0.0, 5.1)
plt.xlim([0, 1])

plt.legend()
```

```python
class LogisticRegressionGD:

    def __init__( self, eta = 0.1, n_iter = 20, random_state = 0 ):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state


    def fit( self, X, y ):
        rgen = np.random.RandomState( self.random_state )
        self.w_ = rgen.normal( loc = 0.0, size = X.shape[1] , scale = 0.1 )
        self.b_ = np.float_(0.)
        self.losses_ = []

        loss = 0
        for i in range( self.n_iter ):
            net_input = self.net_input( X )
            output = self.activation( net_input )
            error = ( y - output )
            self.w_ += 2 * self.eta * X.T.dot(error) / X.shape[0] 
            self.b_ += 2 * self.eta * error / X.shape[0]
            loss = ( np.dot( -y, np.log(output)) - np.dot( ( 1 - y ), np.log( 1 - output ) ) ) / X.shape[0]
            self.losses_.append(loss)
        
        return self
    
    def net_input( self, X ):
        return( np.dot( X, self.w_ ) + self.b_ )
        

    def activation( self, X ):
        return( 1.0 / ( 1 + np.exp( -X.clip(-250, 250) ) ) )

    def predict( self, X ):
        return np.where( self.activation( self.net_input( X ) ) > 0.5, 1, 0 )
```

```python
def plot_decision_regions( X, y, classifier, test_idx = None, res = 0.02 ):

    colors = ("red", "green", "blue", "yellow", "orange", "pink")
    markers = ( "o", "v", "^", "s", "<" )
    cmap = ListedColormap( colors[: len( np.unique(y))] )

    x1_min, x1_max =  X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max =  X[:, 1].min() - 1, X[:, 1].max() + 1

    xx1, xx2 = np.meshgrid( np.arange( x1_min, x1_max, res ),
                            np.arange( x2_min, x2_max, res ))

    prdct = np.array( [xx1.ravel(), xx2.ravel()] ).T
    lab = classifier.predict( prdct )
    lab = lab.reshape( xx1.shape )

    plt.contour( xx1, xx2, lab, alpha = 0.3, cmap = cmap )
    plt.xlim( xx1.min(), xx1.max() )
    plt.ylim( xx2.min(), xx2.max() )

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter( x = X[ y == cl, 0],
                     y = X[ y == cl, 1],
                     alpha = 0.8,
                     c = colors[idx],
                     marker = markers[idx],
                     label = f"Class : {cl}",
                     edgecolor = "black")
        
    if test_idx:
        X_test, Y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0], X_test[:, 1],
        c='none', edgecolor='black', alpha=1.0,
        linewidth=1, marker='o',
        s=100, label='Test set')
```

```python
weights, params = [], []

# The Regression coef and C are reverse proprtional, 1/R = C.
# The more we increase C the less R became and the larger our weights get.
# The more we decrease C the larger R became and the smaller our weights get.

for c in np.arange( -5, 5 ):
    lr = LogisticRegression( C = 10.0**c, multi_class = "ovr" )
    lr.fit( X_std, y )
    weights.append( lr.coef_[1] )
    params.append( 10.0**c )

weights = np.array( weights )
```
